```python
import pandas as pd
from collections import defaultdict
```

```python
# read data from CSV file
df = pd.read_csv("Data_CSV.csv", sep =";")

# convert comman object to float 
def obj_to_float(object):
    if object == "0":
        return int(object)
    else:
        list_obj = list(object)
        i = 0
        while i < len(object):
            if object[i] == ",":
                list_obj[i] = "."
                object = float("".join(list_obj))
                break
            i += 1 
        return object
for row in range(0,63):
    df.loc[row] = df.loc[row].apply(obj_to_float)
```

```python
# create START & END
start = str(input("START: ")).strip().title()
end = str(input("END: ")).strip().title()

# create id for START & END 
for i in range(len(df)):
    if df.iloc[i,0] == start:
        start_id = i
    if df.iloc[i,0] == end:
        end_id = i

print("{} has ID {}.".format(start,start_id))
print("{} has ID {}.".format(end,end_id))
```

```python
# create data for processing
data = defaultdict(list) #create data
over = []

for i in range (1,len(df)): # scan each city to finding neighbor city
    lst = []
    # thêm những thành phố lân cận nếu khoảng cách lớn hơn 0, bé hơn a( ở đây a=4)
    for j in range(1,len(df)):
        if 0 < df.iloc[i,j] < 300: # change 4 -> n after input to check 
            lst.append(df.iloc[0,j])
            lst.append(df.iloc[i,j])
            
    # add h(x)
    lst.append(df.iloc[i,end_id])
    # add to data
    data[str(df.iloc[i][0])] = lst

print(data["Ha Noi"])
```
